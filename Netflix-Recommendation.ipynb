```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_set=pd.read_csv("C://Users//VARSHINI//Downloads//netfix_recommendation.csv",header=None)
transactions=[]
for i in range(0,7465):
    transactions.append([str(data_set.values[i,j]) for j in range(0,20)])
```

```python
print(data_set)
```

```python
from apyori import apriori
intelligence=apriori(transactions=transactions,min_support=0.002,min_confidence=0.2,min_lift=3,min_length=2)
```

```python
results=list(intelligence)
```

```python
results
```

```python
def inspect(results):
    product1=[tuple(result[2][0][0])[0] for result in results]
    product2=[tuple(result[2][0][1])[0] for result in results]
    supports=[result[1] for result in results]
    confidences=[result[2][0][2] for result in results]
    lifts=[result[2][0][3] for result in results]
    return list(zip(product1,product2,supports,confidences,lifts))
DataFrame_intelligence=pd.DataFrame(inspect(results),columns=['Product1','Product2','Support','Confidence','Lift'])
```

```python
print(DataFrame_intelligence)
```

```python
print(DataFrame_intelligence.nlargest(n=20,columns='Lift'))
```
